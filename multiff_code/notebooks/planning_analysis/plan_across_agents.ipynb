```python
import os
import re

def extract_ff_mem(folder):
    basename = os.path.basename(folder)
    match = re.search(r'^ff(\d+)_mem(\d+)_', basename)
    if match:
        return int(match.group(1)), int(match.group(2))
    return float('inf'), float('inf')

```

```python
num_datasets_to_collect = 5


overall_folder_name = "multiff_analysis/RL_models/selected_models/all_agents"
#overall_folder_name = "multiff_analysis/RL_models/sb3_stored_models/all_agents/agents_with_noise/ff5_mem3_drop_fill_visible_only"


pfaa = agent_plan_factors_x_agents_class.PlanFactorsAcrossAgents(overall_folder_name=overall_folder_name)


# ff1_agent_folders = [
#     folder for folder in pfaa.agent_folders
#     if '/ff1_' in folder
# ]
# pfaa.agent_folders = ff1_agent_folders

sorted_agent_folders = sorted(
    pfaa.agent_folders,
    key=extract_ff_mem
)
pfaa.agent_folders = sorted_agent_folders

pfaa.make_all_ref_pooled_median_x_agents_AND_pooled_perc_x_agents(exists_ok=False,
                                                                num_datasets_to_collect=num_datasets_to_collect,
                                                                intermediate_products_exist_ok=True,
                                                                agent_data_exists_ok=True,
                                                                use_stored_data_only=use_stored_data_only
                                                                )
all_ref_pooled_median_x_agents = pfaa.all_ref_pooled_median_x_agents.copy()
pooled_perc_x_agents = pfaa.pooled_perc_x_agents.copy()
```
