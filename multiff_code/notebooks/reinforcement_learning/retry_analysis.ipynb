```python
all_agents_ids = [
#  "ff3_mem3_rank_keep_visible_and_memory_job201550_22",
#  "ff2_mem3_rank_keep_visible_and_memory_job201550_14",
 "ff3_mem3_rank_keep_visible_only_job201550_21",
#  "ff2_mem3_rank_keep_visible_only_job201550_13",   
"ff3_mem3_drop_fill_visible_only_job201550_23",
]
```

```python
all_agents_pattern_frequencies = pd.DataFrame()

for agent_id in all_agents_ids:
    print(f'Processing agent {agent_id}')
    model_folder_name = f"multiff_analysis/RL_models/selected_models/all_agents/{agent_id}"
    agent = sb3_class.SB3forMultifirefly(model_folder_name=model_folder_name)
    agent.streamline_getting_data_from_agent(
        n_steps=9000, exists_ok=True, save_data=True)
    agent.ff_dataframe = make_ff_dataframe.furnish_ff_dataframe(agent.ff_dataframe, agent.ff_real_position_sorted,
                                                                agent.ff_caught_T_new, agent.ff_life_sorted)
    agent.make_df_related_to_patterns_and_features()
    agent.pattern_frequencies['id'] = agent_id
    
    all_agents_pattern_frequencies = pd.concat([all_agents_pattern_frequencies, agent.pattern_frequencies], axis=0)

all_agents_pattern_frequencies.reset_index(drop=True, inplace=True)




```
