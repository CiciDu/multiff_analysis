```python
# changeable_variables=['opt_arc_type', 'ref_point_value']
changeable_variables=[]

fixed_variable_values_to_use={'opt_arc_type': 'opt_arc_stop_closest',
                                'ref_point_value': -100,
                                }


agent_median_df2 = agent_median_df.copy()
for col in ['diff_in_abs_angle_to_nxt_ff_median', 'diff_in_abs_angle_to_nxt_ff_ci_low_95', 'diff_in_abs_angle_to_nxt_ff_ci_high_95']:
    agent_median_df2[col] = agent_median_df2[col] * 180 / np.pi


for sweep_key in ['obs_mem_r', 'obs_mem_th']:

    subset, fixed_values = plot_agent_patterns.build_conditioned_subset(
        agent_median_df2,
        fixed_values=plot_agent_patterns.build_fixed_values_for_obs_sweep(
            sweep_key=sweep_key,
        ),
    )
    # heading (default)
    plot_agent_patterns.streamline_plot_agents_with_conditions(
        subset,
        sweep_key=sweep_key,
        changeable_variables=changeable_variables,
        fixed_variable_values_to_use=fixed_variable_values_to_use,
    )
    plt.show()

    # curv
    if 'diff_in_abs_d_curv_median' in subset.columns:
        plot_agent_patterns.streamline_plot_agents_with_conditions(
            subset,
            sweep_key=sweep_key,
            changeable_variables=changeable_variables,
            fixed_variable_values_to_use=fixed_variable_values_to_use,
            y_key='diff_in_abs_d_curv_median',
        )
        plt.show()

    # perc (from agent_perc_df)
    perc_subset, _ = plot_agent_patterns.build_conditioned_subset(
        agent_perc_df,
        fixed_values=fixed_values,
    )
    if perc_subset is not None and 'perc' in perc_subset.columns and sweep_key in perc_subset.columns:
        plot_agent_patterns.streamline_plot_agents_with_conditions(
            perc_subset,
            sweep_key=sweep_key,
            changeable_variables=[c for c in changeable_variables if c in perc_subset.columns],
            y_key='perc',
        )
        plt.show()
```
