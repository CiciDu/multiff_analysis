## Loop over retry events specifically

```python
# ── pick one agent from the sweep ────────────────────────────────────────
model_folder_name = 'RL_models/sb3_stored_models/all_agents/agents_with_noise/mr0p0_mt0p5_seed6_rep4'
pa = agent_plan_factors_class.PlanFactorsOfAgent(model_folder_name=model_folder_name)
pa.get_agent_data(n_steps=8000, exists_ok=True, save_data=True)

# ── build stop_category_df, then extract retry (rsw) events ──────────────
pa.agent.make_or_retrieve_stop_category_df()
rsw_events_df, _ = detect_rsw_and_rcap.make_rsw_events_df(
    pa.agent.stop_category_df,
    pa.agent.ff_real_position_sorted,
    pa.agent.monkey_information,
)
print(f"{len(rsw_events_df)} retry (rsw) events found")

# ── plot each retry event ─────────────────────────────────────────────────
lead_time  = 3   # seconds before first stop
trail_time = 3   # seconds after last stop

n_cols = 3
n_rows = int(np.ceil(len(rsw_events_df) / n_cols))
fig, axes_grid = plt.subplots(n_rows, n_cols, figsize=(6 * n_cols, 6 * n_rows))
axes_flat = np.array(axes_grid).flatten()

for i, (_, row) in enumerate(rsw_events_df.iterrows()):
    t_start = row['stop_1_time'] - lead_time
    t_end   = row['last_stop_time'] + trail_time

    plot_trials.PlotTrials(
        duration=[t_start, t_end],
        monkey_information=pa.agent.monkey_information,
        ff_dataframe=pa.agent.ff_dataframe,
        ff_life_sorted=pa.agent.ff_life_sorted,
        ff_real_position_sorted=pa.agent.ff_real_position_sorted,
        ff_believed_position_sorted=pa.agent.ff_believed_position_sorted,
        cluster_around_target_indices=[],
        ff_caught_T_new=pa.agent.ff_caught_T_new,
        player='monkey',
        show_stops=True,
        show_alive_fireflies=True,
        fig=fig,
        axes=axes_flat[i],
    )
    axes_flat[i].set_title(
        f"rsw #{i}  trial={row['trial']}  nstops={row['num_stops']}"
    )

for j in range(i + 1, len(axes_flat)):
    axes_flat[j].set_visible(False)

plt.tight_layout()
plt.show()
```
