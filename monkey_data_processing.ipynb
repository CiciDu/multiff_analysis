# Make animation

```python
dir_name = 'all_monkey_data/individual_monkey_data'
num_trials = 3
k = 3
#for monkey_name in os.listdir(dir_name):
for monkey_name in ['monkey_Quigley']: #'monkey_Schro', 'monkey_Bruno']:
    monkey_path = os.path.join(dir_name, monkey_name)
    for data_name in os.listdir(monkey_path):
        temp_raw_data_folder_name = os.path.join(monkey_path, data_name)
        data_instance = data_processing_class.ProcessMonkeyData(raw_data_folder_name=temp_raw_data_folder_name)
        print(temp_raw_data_folder_name)
        
        animation_folder_name = os.path.join('animation', monkey_name, data_name)

        if not exists(animation_folder_name):
            os.makedirs(animation_folder_name, exist_ok = True)   

        if len(os.listdir(animation_folder_name)) < 8: 
            # Then we need to save some more animations;
            # We use this condition so as not to repeat what we've already done
            data_instance.retrieve_monkey_data()
            data_instance.make_or_retrieve_ff_dataframe()
            # Prepare for animatoin
            sampled_trials = np.round(np.linspace(num_trials, len(data_instance.ff_caught_T_sorted)-1, 7)).astype('int')
            for currentTrial in sampled_trials:
                data_instance.set_animation_parameters(currentTrial, num_trials, k)
                data_instance.make_animation(margin=400, save_video=False)
                
                # Save the animation
                writervideo = animation.FFMpegWriter(fps=int(62/k))
                filename = "trial_" + str(currentTrial-data_instance.new_num_trials+1) + "_to_" + str(currentTrial)
                data_instance.anim.save(f"{animation_folder_name}/{filename}.mp4", writer=writervideo)
                print(f"Saved {animation_folder_name}/{filename}.mp4")
```
