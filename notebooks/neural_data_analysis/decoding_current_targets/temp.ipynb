```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_lag_offset_train_test_overlap(
    df, dataset_name, chunk_size=30, alpha=0.8,
    base_width=0.35, narrower_width_ratio=0.6,
    width_per_var=0.6, height=6,
    mode='train_offset'  # 'train_offset' or 'lag_offset'
):
    """
    Plot bars with offset and width controlled by either
    train/test vs lag/no_lag groups.

    Parameters:
        df: DataFrame with columns ['var', 'whether_lag', 'train_or_test', 'corr']
        dataset_name: str for title
        chunk_size: how many variables per plot
        alpha: transparency of bars
        base_width: base bar width
        narrower_width_ratio: ratio for the narrower bars
        width_per_var: figure width scaling per variable
        height: figure height
        mode: 
            'train_offset' - train/test offset, lag width (default)
            'lag_offset' - lag offset, train/test width
    """

    vars_unique = df['var'].unique()
    num_chunks = int(np.ceil(len(vars_unique) / chunk_size))

    train_test_statuses = ['train', 'test']
    lag_statuses = sorted(df['whether_lag'].unique())  # e.g., ['lag', 'no_lag']

    if mode == 'train_offset':
        # train/test controls offset, lag controls width
        offset_groups = train_test_statuses
        width_groups = lag_statuses
        colors = {
            ('train', 'no_lag'): '#1f77b4',      # Blue (train + lag)
            ('train', 'lag'): '#aec7e8',   # Light blue (train + no lag)
            ('test', 'no_lag'): '#ff7f0e',       # Orange (test + lag)
            ('test', 'lag'): '#ffbb78',    # Light orange (test + no lag)
        }
    elif mode == 'lag_offset':
        # lag controls offset, train/test controls width
        offset_groups = lag_statuses
        width_groups = train_test_statuses
        colors = {
            ('lag', 'train'): '#1f77b4',    # Blue
            ('lag', 'test'): '#d62728',     # Red
            ('no_lag', 'train'): '#2ca02c', # Green
            ('no_lag', 'test'): '#ff7f0e',  # Orange
        }
    else:
        raise ValueError("mode must be 'train_offset' or 'lag_offset'")

    for i in range(num_chunks):
        chunk_vars = vars_unique[i * chunk_size:(i + 1) * chunk_size]
        chunk_df = df[df['var'].isin(chunk_vars)].copy()
        chunk_df['var'] = pd.Categorical(chunk_df['var'], categories=chunk_vars, ordered=True)

        fig_width = max(chunk_size * width_per_var, 10)
        fig, ax = plt.subplots(figsize=(fig_width, height))

        x = np.arange(len(chunk_vars))
        n_offsets = len(offset_groups)
        offset_width = base_width * 1.5  # spacing between offset groups

        for idx, offset_group in enumerate(offset_groups):
            offset = (idx - (n_offsets - 1) / 2) * offset_width

            for width_group in width_groups:
                # Filter df by offset and width groups depending on mode
                if mode == 'train_offset':
                    subset_df = chunk_df[
                        (chunk_df['train_or_test'] == offset_group) &
                        (chunk_df['whether_lag'] == width_group)
                    ].set_index('var').reindex(chunk_vars)
                    width = base_width if width_group == 'lag' else base_width * narrower_width_ratio
                    color = colors.get((offset_group, width_group), 'gray')
                    alpha_adj = alpha
                else:  # lag_offset
                    subset_df = chunk_df[
                        (chunk_df['whether_lag'] == offset_group) &
                        (chunk_df['train_or_test'] == width_group)
                    ].set_index('var').reindex(chunk_vars)
                    width = base_width if width_group == 'train' else base_width * narrower_width_ratio
                    color = colors.get((offset_group, width_group), 'gray')
                    alpha_adj = alpha if width_group == 'train' else alpha * 0.7

                ax.bar(
                    x + offset, subset_df['corr'],
                    width=width,
                    alpha=alpha_adj,
                    label=f'{offset_group.capitalize()} - {width_group}',
                    color=color
                )

        ax.set_title(f"{dataset_name} - Variables {i * chunk_size + 1} to {(i + 1) * chunk_size}")
        ax.set_xticks(x)
        ax.set_xticklabels(chunk_vars, rotation=45, ha='right')
        ax.set_xlabel("Variable")
        ax.set_ylabel("Canonical Correlation")
        
        ax.axhline(0.1, color='purple', linestyle='--', linewidth=2, alpha=0.5)

        # Add vertical dashed grid lines separating variables
        for grid_pos in x[:-1] + 0.5:
            ax.axvline(grid_pos, color='gray', linestyle='--', linewidth=0.7, alpha=0.5)

        ax.grid(False)
        ax.yaxis.grid(True, linestyle=':', alpha=0.7)

        # Remove duplicate legend entries
        handles, labels = ax.get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        ax.legend(by_label.values(), by_label.keys(), title='Groups', loc='best')

        plt.tight_layout()
        plt.show()

        print("=" * 150)

plot_lag_offset_train_test_overlap(combined_X2_df, 'DatasetName', mode='train_offset')
```
